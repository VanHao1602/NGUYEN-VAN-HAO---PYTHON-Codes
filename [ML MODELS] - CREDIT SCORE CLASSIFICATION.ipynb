```python
def filter_col(value): #lọc giá trị trong một chuỗi dựa trên một số tiêu chí
    if '-' in value: #Nếu chuỗi chứa dấu "-", hàm sẽ trả về phần sau của chuỗi sau khi tách thành hai phần dựa trên dấu "-".
        return value.split('-')[1]
    elif '_' in value: # Nếu chuỗi chứa dấu "", hàm sẽ trả về phần trước của chuỗi sau khi tách thành hai phần dựa trên dấu "".
        return value.split('_')[0]
    else: # Trong các trường hợp còn lại, hàm sẽ trả về chính chuỗi đầu vào.
        return value
```

```python
df['Age'] = df['Age'].apply(filter_col) # lọc giá trị chuỗi loại bỏ "-" và "_"
```

```python
# đổi về dạng số 
def parse_years_and_months(age):
    if isinstance(age, str):
        age_parts = age.split(' Years and ')
        years = int(age_parts[0]) if 'Years' in age else 0
        months_str = age_parts[1].split(' Months')[0] if 'Months' in age_parts[1] else '0'
        months = int(months_str)
        total_months = years * 12 + months
        return total_months
    else:
        return 0
df['Credit_History_Age_Months'] = df['Credit_History_Age'].apply(parse_years_and_months)
df.drop('Credit_History_Age',axis=1,inplace=True)
```

```python
df[['Credit_History_Age_Months', 'Customer_ID','Month']].head(16) # Những giá trị còn thiếu đều có quy luật nên ta xây dựng hàm để điền các giá trị NaN trong cột 'Credit_History_Age_Months'
```

```python
from tqdm import tqdm

def fill_nan_credit_history_age(df):
    # Sắp xếp dữ liệu theo 'Customer_ID' và 'Month'
    df.sort_values(by=['Customer_ID', 'Month'], inplace=True)
    
    # Khởi tạo tiến trình tqdm để theo dõi tiến độ
    progress_bar = tqdm(total=df.shape[0])
    
    # Lặp qua từng 'Customer_ID' trong bảng dữ liệu
    for customer_id in df['Customer_ID'].unique():
        # Lấy các dòng có 'Customer_ID' tương ứng
        customer_rows = df[df['Customer_ID'] == customer_id]
        
        # Tìm giá trị đầu tiên của 'Credit_History_Age_Months' không phải NaN
        first_valid_index = customer_rows['Credit_History_Age_Months'].first_valid_index()
        
        # Nếu không có giá trị đầu tiên không phải NaN, bỏ qua
        if first_valid_index is None:
            continue
        
        # Lấy giá trị 'Credit_History_Age_Months' đầu tiên không phải NaN
        first_valid_value = customer_rows.loc[first_valid_index, 'Credit_History_Age_Months']
        
        # Tính toán giá trị cho các giá trị NaN
        current_value = first_valid_value
        for index, row in customer_rows.iterrows():
            if pd.isnull(row['Credit_History_Age_Months']):
                df.at[index, 'Credit_History_Age_Months'] = current_value
            else:
                current_value = row['Credit_History_Age_Months']
            current_value += 1
        
            # Cập nhật tiến độ của tqdm
            progress_bar.update(1)
    
    # Đóng tiến trình tqdm
    progress_bar.close()
    
    return df

# Sử dụng hàm để điền các giá trị NaN trong cột 'Credit_History_Age_Months'
df_filled = fill_nan_credit_history_age(df.copy())

# In ra DataFrame sau khi điền giá trị NaN
print(df_filled)
```

Tiến hành tunning mô hình bằng cách:

- Dùng vòng lặp thay đổi tham số max_depth chạy từ 2 đến 16
- Thay đổi lần lượt cùng với criteria entropy và gini


--------------------------------------------
Perform model tuning by:

- Using a loop to change the `max_depth` parameter from 2 to 16.
- Alternating between criteria, using both "entropy" and "gini".


```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
import seaborn as sns

# Khởi tạo biến để lưu trữ mô hình tốt nhất và độ chính xác tốt nhất
best_model = None
best_accuracy = 0

# List of different criterion
criterion = ['gini', 'entropy']

for c in criterion:
    # Fit Decision Tree classifier
    clf = DecisionTreeClassifier(criterion=c, random_state=0, max_depth=9)
    clf.fit(X_train_smote, y_train_smote)
    y_pred_tree = clf.predict(X_valid)

    # Tính độ chính xác
    accuracy = accuracy_score(y_pred_tree, y_valid)

    # Lưu trữ mô hình tốt nhất và độ chính xác tốt nhất
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_model = clf

    # Báo cáo confusion matrix
    print("Criterion:", c)
    print('Accuracy: ', accuracy)
    print(classification_report(y_valid, y_pred_tree))

# Vẽ confusion matrix cho mô hình tốt nhất
y_pred_best = best_model.predict(X_valid)
cm_best = confusion_matrix(y_valid, y_pred_best)
plt.figure(figsize=(6, 6))
sns.heatmap(cm_best, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Confusion Matrix for Best Model')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()
```

Kết quả trả ra hiệu suất tốt nhất (Accuracy Max) tại `max_depth` = 9, criteria là gini cụ thể như sau:

- F1-score đối với class 1,2,3 lần lượt là 0.7; 0.73; 0.64
- Precision đối với class 1,2,3 lần lượt là 0.71; 0.69; 0.73
- Recall đối với class 1,2,3 lần lượt là 0.69; 0.78; 0.57
- Accuracy là 0.7

--> Có thể thấy dù ít nhưng hiệu suất mô hình sau khi tunning đã có cải thiện so với mô hình ban đầu

-------------------------------------------------
The best performance result (Maximum Accuracy) is achieved at `max_depth` = 9 with the "gini" criterion, as follows:

- F1-score for class 1, 2, 3 are 0.7, 0.73, 0.64 respectively.
- Precision for class 1, 2, 3 are 0.71, 0.69, 0.73 respectively.

After the data processing, cleaning, exploration, and model building, we can draw some conclusions as follows:

- The number of customers and credit scores are fairly evenly distributed across industries.
- The age of customers is primarily between 20 and 50 years old, and customers aged 40 to 50 tend to have better credit scores than those under 30 (due to better financial capacity and more experience).
- Customers with a better credit mix tend to have better credit scores. This index reflects diversity and balance in a customer's financial management, which reduces the risk of credit defaults.
- A customer with a higher number of bank accounts and credit cards is usually rated higher on their credit score, as this allows the bank to better control the customer’s cash flow. To some extent, having many accounts and credit cards may indicate potential money laundering behavior.
- The fewer loans a customer has and the fewer overdue payments they have, the higher their credit score will be, reflecting better financial capacity and the ability to manage payments.
- Customers with good credit scores are typically offered preferential interest rates to encourage spending, whereas customers with low credit scores are charged higher interest rates due to previous late payment increases.
